```python
import cv2 as cv
import torch
from torchvision import transforms
from Model import DeePixBiS

# Load the model
model = DeePixBiS()
model.load_state_dict(torch.load('./DeePixBiS.pth'))
model.eval()

# Transformation for input images
tfms = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Face classifier
faceClassifier = cv.CascadeClassifier('Classifiers/haarface.xml')

def predict_face(image_path):
    # Read image
    img = cv.imread(image_path)
    grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    # Perform face detection
    faces = faceClassifier.detectMultiScale(grey, scaleFactor=1.1, minNeighbors=4)

    # Prepare results dictionary
    results = {}

    # Iterate through detected faces
    for i, (x, y, w, h) in enumerate(faces):
        faceRegion = img[y:y + h, x:x + w]
        faceRegion = cv.cvtColor(faceRegion, cv.COLOR_BGR2RGB)

        # Apply transformations
        faceRegion = tfms(faceRegion)
        faceRegion = faceRegion.unsqueeze(0)

        # Perform classification
        mask, _ = model.forward(faceRegion)
        res = torch.mean(mask).item()
        prediction = 'fake' if res < 0.5 else 'real'

        # Store prediction for the face
        results[f'Face_{i+1}'] = prediction

    return results

# Example usage:
image_path = '/content/Face-Anti-Spoofing-using-DeePixBiS-main/data/images/fake_0001_00_00_01_0.jpg'
predictions = predict_face(image_path)
print(predictions)
```
